## Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
train_data, test_data = train_test_split(data, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
train_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['temp', 'atemp', 'hum', 'windspeed', 'casual','registered','cnt']

train_data[num_vars] = scaler.fit_transform(train_data[num_vars])
```

```python
y_train = train_data.pop('cnt')
x_train = train_data
```

```python
import statsmodels.api as sm

train_lm = sm.add_constant(x_train)
```

```python
## Checking the VIF

from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

### Model Building Using the RFE method

```python
# Importing RFE and LinearRegression

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()

lm.fit(x_train,y_train)


rfe = RFE(lm,6)
rfe = rfe.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = x_train.columns[rfe.support_]
col
```

```python
x_train_rfe = x_train[col]
```

```python
x_train_rfe = sm.add_constant(x_train_rfe)
```

```python
lm_rfe = sm.OLS(y_train,x_train_rfe).fit()
```

```python
print(lm_rfe.summary())
```

```python
 # Dropping highly correlated variables and insignificant variables

x_train_rfe = x_train_rfe.drop('Summer', 1)
```

```python
x_train_lm = sm.add_constant(x_train_rfe)
```

```python
lm_rfe1 = sm.OLS(y_train,x_train_lm).fit()
```

```python
print(lm_rfe1.summary())
```
